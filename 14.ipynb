```python
import numpy as np
from PIL import Image

def average_images(image_list):
    totalArray = np.array(Image.open(image_list[0]), 'f')
    count=1

    for imagePath in image_list[1:]:
        try:
            imageArray = np.array(Image.open(imagePath), 'f')
            totalArray += imageArray
            count=count+1
        except:
            print("Skip ", imagePath);

    # Tính trung bình
    averageArray = totalArray / count

    # Chuyển đổi kết quả trung bình thành hình ảnh
    averageImage = Image.fromarray(averageArray.astype('uint8'))

    return averageImage
```

```python
# Ví dụ 1:
imagePath1 = 'C:\\Users\\wwwkb\\OneDrive\\Desktop\\458333546_904212478415207_1207610342633984991_n.jpg'
imagePath2 = 'C:\\Users\\wwwkb\\OneDrive\\Desktop\\z5716685533800_3f5017b98aec6682633b292458189613.jpg'
image_list = [imagePath1, imagePath2]
resultImage = average_images(image_list)
display(resultImage)


# Ví dụ 2:
imagePath3 = 'C:\\Users\\wwwkb\\OneDrive\\Desktop\\Đề tài thị giác máy tính.png'
imagePath4 = 'C:\\Users\\wwwkb\\OneDrive\\Desktop\\Untitlesd.png'
image_list = [imagePath3, imagePath4]
resultImage = average_images(image_list)
display(resultImage)
```
