```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

image = Image.open('C:\\Users\\wwwkb\\OneDrive\\Desktop\\istockphoto-1176352693-612x612.jpg')

# Chuyển sang ảnh xám
grayImage = image.convert('L')

imageArray = np.array(grayImage)

# Chuyển sang ảnh nhị phân
binaryImage = 1*(imageArray<200)

# Đánh nhãn
labeledImage, numFeatures = ndimage.label(binaryImage)

# Tạo lưới subplot với 3 dòng và 1 cột
fig, axs = plt.subplots(3, 1, figsize=(50, 50))

plt.subplot(3,1,1)
plt.imshow(grayImage, cmap='gray')
plt.title("Ảnh xám")

plt.subplot(3,1,2)
plt.imshow(binaryImage, cmap='gray')
plt.title("Ảnh nhị phân")

plt.subplot(3,1,3)
plt.imshow(labeledImage, cmap='gray')
plt.title(f"Đã đánh nhãn ({numFeatures} đối tượng)")

plt.show();
```
