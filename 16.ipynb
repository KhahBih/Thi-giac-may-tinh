```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

# Đọc ảnh
imagePath = 'C:\\Users\\wwwkb\\OneDrive\\Desktop\\458548812_434784309612677_8960315865400473921_n.jpg'
image = Image.open(imagePath)

# Ảnh xám
grayImage = image.convert('L')

# Chuyển ảnh xám về mảng
image_a = np.array(grayImage)

# Làm mờ ảnh xám với bộ lọc Gaussian
blurredImageA = ndimage.gaussian_filter(image_a, sigma=1, mode="constant")
blurredImageB = ndimage.gaussian_filter(image_a, sigma=3, mode="constant")
blurredImageC = ndimage.gaussian_filter(image_a, sigma=5, mode="constant")

# Hiển thị ảnh
fig, axs = plt.subplots(2,2, figsize=(10, 8))

plt.subplot(2,2,1)
plt.imshow(grayImage, cmap='gray')
plt.title('Ảnh xám gốc')

plt.subplot(2,2,2)
plt.imshow(blurredImageA, cmap='gray')
plt.title('sigma=1')

plt.subplot(2,2,3)
plt.imshow(blurredImageB, cmap='gray')
plt.title('sigma=3')

plt.subplot(2,2,4)
plt.imshow(blurredImageC, cmap='gray')
plt.title('sigma=5')
```
