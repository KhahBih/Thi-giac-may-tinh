```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
imagePath = 'C:\\Users\\wwwkb\\OneDrive\\Desktop\\test.jpg'
```

```python
#c nh
image = Image.open(imagePath)
plt.imshow(image)
```

```python
#Chuyn nh sang mu xm
imageGray = image.convert('L')
plt.imshow(imageGray, cmap='gray')
```

```python
imageArray = np.array(imageGray)
print(imageArray)
```

```python
# To mt hnh v Histogram
plt.figure()
plt.hist(imageArray.flatten(), bins=128)
plt.title('Biu  Histogram ca nh')
plt.xlabel('Gi tr ca pixel')
plt.ylabel('S lng')
plt.show()
```
